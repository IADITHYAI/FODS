```python
q1=data['Salary'].quantile(0.25)
q2=data['Salary'].quantile(0.75)
```

```python
IQR=q2-q1
lb=q1-1.5*IQR
ub=q2+1.5*IQR
```

```python
outlier=(data[['Salary']]<lb) | (data[['Salary']]>ub)
```

```python
cleaned_data=data[~outlier]
```

```python
cleaned_data
```

```python
minn=cleaned_data['Salary'].min()
maxx=cleaned_data['Salary'].max()
```

```python
cleaned_data['Salary']=(cleaned_data['Salary']-minn)/(maxx-minn)
cleaned_data
```

```python
dummies=pd.get_dummies(cleaned_data,columns=['Location'],drop_first=True)
dummies
```
